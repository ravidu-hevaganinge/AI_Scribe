```python
import numpy as np
import matplotlib.pyplot as plt

# Averages
rouge1_avg = [37.059, 46.637]
rouge2_avg = [22.8755, 21.7576]
rougeL_avg = [33.762, 27.002]

# Categories
categories = ['Initial Results', 'Most Recent Results']
bar_width = 0.25
index = np.arange(len(categories))

# Define pastel colors
pastel_red = (1.0, 0.6, 0.6)
pastel_green = (0.6, 1.0, 0.6)
pastel_blue = (0.6, 0.6, 1.0)

# Plotting ROUGE-1
plt.bar(index, rouge1_avg, bar_width, label='ROUGE-1', color=pastel_red)
# Plotting ROUGE-2
plt.bar(index + bar_width, rouge2_avg, bar_width, label='ROUGE-2', color=pastel_green)
# Plotting ROUGE-L
plt.bar(index + 2*bar_width, rougeL_avg, bar_width, label='ROUGE-L', color=pastel_blue)

# Adding values on top of bars
for i in range(len(categories)):
    plt.text(index[i] - 0.02, rouge1_avg[i] + 0.2, str(rouge1_avg[i]), color='black')
    plt.text(index[i] + bar_width - 0.02, rouge2_avg[i] + 0.2, str(rouge2_avg[i]), color='black')
    plt.text(index[i] + 2*bar_width - 0.02, rougeL_avg[i] + 0.2, str(rougeL_avg[i]), color='black')

plt.xlabel('Categories')
plt.ylabel('ROUGE Scores')
plt.title('Comparison of Initial Results vs Most Recent Results')
plt.xticks(index + bar_width, categories)
plt.legend()
plt.tight_layout()
plt.show()
```
