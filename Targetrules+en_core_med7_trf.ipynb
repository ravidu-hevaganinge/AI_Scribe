```python
import pandas as pd
import spacy
from spacy.tokens import Span
from medspacy.target_matcher import TargetRule, TargetMatcher
```

```python
import medspacy
#from medspacy.ner import TargetRule

Span.set_extension("icd10", default="", force=True)
Span.set_extension("category", default="", force=True)
target_matcher = TargetMatcher(nlp1)

csv_file = "/content/Capstone Symptom Data Collection - Sheet1.csv"
df1 = pd.read_csv(csv_file)

csv_file = "/content/medications_extended.csv"
df2 = pd.read_csv(csv_file)

csv_file = "/content/Capstone Location of body Data Collection - Sheet1 (1).csv"
df3 = pd.read_csv(csv_file)

csv_file = "/content/Quality  - Sheet1.csv"
df4 = pd.read_csv(csv_file)

csv_file = "/content/Severity  - Sheet1 (1).csv"
df5 = pd.read_csv(csv_file)

# Define target rules for each symptom
target_rules = []
for symptom in df1["Symptoms"]:
    pattern = [{"LOWER": {"IN": [word.lower() for word in symptom.split()]}}]
    target_rule = TargetRule(symptom, category="SYMPTOM", pattern=pattern)
    target_rules.append(target_rule)

for medication in df2["GenericName"]:
    pattern = [{"LOWER": {"IN": [word.lower() for word in medication.split()]}}]
    target_rule = TargetRule(medication, category="MEDICATION", pattern=pattern)
    target_rules.append(target_rule)

for location in df3["Location"]:
    pattern = [{"LOWER": {"IN": [word.lower() for word in location.split()]}}]
    target_rule = TargetRule(location, category="LOCATION", pattern=pattern)
    target_rules.append(target_rule)

for quality in df4["Quality "]:
    pattern = [{"LOWER": {"IN": [word.lower() for word in location.split()]}}]
    target_rule = TargetRule(quality, category="QUALITY", pattern=pattern)
    target_rules.append(target_rule)

for severity in df5["Severity "]:
    pattern = [{"LOWER": {"IN": [word.lower() for word in location.split()]}}]
    target_rule = TargetRule(severity, category="SEVERITY", pattern=pattern)
    target_rules.append(target_rule)

target_matcher.add(target_rules)
```

```python
nlp1.get_pipe('medspacy_target_matcher').add(target_rules)

doc = nlp1('''Physician: Good morning, Mrs. Smith. How can I assist you today?

Patient (Mrs. Smith): Good morning, Doctor. I've been experiencing some discomfort lately.

Physician: I see. Can you tell me more about the pain you're feeling?

Mrs. Smith: Yes, it's a sharp pain in my lower back.

Physician: Okay, and on a scale of 1 to 10, how would you rate the severity of this pain?

Mrs. Smith: I would say it's about a 7.

Physician: How long have you been experiencing this pain?

Mrs. Smith: It started about a week ago.

Physician: Do you notice any patterns in when the pain occurs?

Mrs. Smith: It tends to get worse when I'm sitting for long periods or when I try to lift something heavy and its severe.

Physician: Are there any activities or movements that seem to trigger the pain?

Mrs. Smith: Yes, it usually starts when I'm bending down to pick something up.

Physician: I see. And have you noticed anything that makes the pain better or worse?

Mrs. Smith: Taking over-the-counter pain relievers like aspirin seems to help a bit.

Physician: Thank you for sharing that. Could you please provide details about the aspirin you're taking? This includes the dosage, frequency, route, and any other relevant information.

Mrs. Smith: Sure. I take 2 tablets of 325 mg aspirin orally every 6 hours as needed.

Physician: Alright, thank you. Are there any other medications you're currently taking?

Mrs. Smith: No, just the aspirin for now.

Physician: Okay, let's move on. Are there any other symptoms you've been experiencing along with the pain?

Mrs. Smith: Sometimes I feel a tingling sensation down my leg, but it's not constant.

Physician: Thank you for sharing that. Do you have any relevant medical history or family history of similar issues?

Mrs. Smith: Well, my mother had some back problems when she was older, but nothing too serious.

Physician: Alright, thank you for providing that information. Based on what you've told me, it sounds like you may be experiencing some nerve compression in your lower back. I'd like to conduct a physical examination to confirm this and discuss potential treatment options. Does that sound okay with you?

Mrs. Smith: Yes, Doctor, that sounds fine.

''')
for ent in doc.ents:
    print(ent, ent._.is_negated, ent._.is_family, ent._.is_historical)
medspacy.visualization.visualize_ent(doc)
```
