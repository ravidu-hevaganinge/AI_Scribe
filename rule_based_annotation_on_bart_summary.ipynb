```python
# Access the target matcher pipe
target_matcher = nlp1.get_pipe("medspacy_target_matcher")

# Define extensions
Span.set_extension("icd10", default="", force=True)
Span.set_extension("category", default="", force=True)

# Read dataframes
csv_files = ["/content/drive/MyDrive/UMD BIOE CAPSTONE PROJECT/NER/NER_CSV_FILES/Capstone Symptom Data Collection - Sheet1.csv",
             "/content/drive/MyDrive/UMD BIOE CAPSTONE PROJECT/NER/NER_CSV_FILES/medications_extended.csv",
             "/content/drive/MyDrive/UMD BIOE CAPSTONE PROJECT/NER/NER_CSV_FILES/Capstone Location of body Data Collection - Sheet1 (1).csv",
             "/content/drive/MyDrive/UMD BIOE CAPSTONE PROJECT/NER/NER_CSV_FILES/Quality  - Sheet1.csv",
             "/content/drive/MyDrive/UMD BIOE CAPSTONE PROJECT/NER/NER_CSV_FILES/Severity  - Sheet1 (1).csv"]

dfs = [pd.read_csv(file) for file in csv_files]

# Define target rules for each category
target_rules = []
for df, category in zip(dfs, ["SYMPTOM", "MEDICATION", "LOCATION", "QUALITY", "SEVERITY"]):
    for term in df.iloc[:, 0]:  # Assuming the first column contains the terms
        pattern = [{"LOWER": {"IN": [word.lower() for word in term.split()]}}]
        target_rule = TargetRule(term, category=category, pattern=pattern)
        target_rules.append(target_rule)

# Add target rules to target matcher
target_matcher.add(target_rules)
```

```python
import pandas as pd

# Function to convert spaCy doc object to pandas DataFrame
def convert_doc_to_pandas(doc_object):
    # Initialize lists to hold extracted data
    data = {
        "Text": [],
        "Start": [],
        "End": [],
        "Label": []
    }

    # Extract entities related to drug, route, strength, and dosage
    for ent in doc_object.ents:
        if ent.label_ in ["STRENGTH","DRUG", "FREQUENCY"]:
            data["Text"].append(ent.text)
            data["Start"].append(ent.start_char)
            data["End"].append(ent.end_char)
            data["Label"].append(ent.label_)

    dataframe = pd.DataFrame(data)
    return dataframe

# Convert annotations to DataFrames
transcript_annotations_df = convert_doc_to_pandas(doc_0)
summary_hpi_annotations_df = convert_doc_to_pandas(doc_1)

# Compare annotations between transcript and summary HPI
annotations_in_transcript_not_in_summary = transcript_annotations_df[~transcript_annotations_df['Text'].isin(summary_hpi_annotations_df['Text'])]

# Display DataFrames and missing annotations
print("Transcript Annotations:")
print(transcript_annotations_df)

print("\nSummary HPI Annotations:")
print(summary_hpi_annotations_df)

print("\nAnnotations present in transcript but not in Summary HPI:")
print(annotations_in_transcript_not_in_summary)

```

```python
# Function to generate template sentences for missing annotations and enhance HPI
def enhance_summary_with_missing_annotations(transcript_annotations_df, generated_summary):
    # Generate template sentences based on missing annotations
    sample_sentences = {
        "STRENGTH": " ({0})",
        "DRUG": " The patient reports taking {0}",
        "FREQUENCY": "{0}."
    }

    # Convert generated summary to a list of words
    generated_summary_words = generated_summary.split()
    additional_sentences_words = []
    added_sentences = []  # List to store added sentences

    # Add words from additional sentences to the list
    for label, text in zip(transcript_annotations_df["Label"], transcript_annotations_df["Text"]):
        if label in sample_sentences:
            # Generate the additional sentence
            additional_sentence = sample_sentences[label].format(text, '')
            added_sentences.append(additional_sentence)
            # Split the formatted sentence into words and add to the list
            additional_sentences_words.extend(additional_sentence.split())

    # Concatenate the lists of words
    enhanced_summary_words = generated_summary_words + additional_sentences_words

    # Join the words back into sentences
    enhanced_summary = ' '.join(enhanced_summary_words)

    return enhanced_summary, added_sentences

# Call the function to enhance generated summary
enhanced_summary, added_sentences = enhance_summary_with_missing_annotations(transcript_annotations_df, generated_summary)

# Print original generated summary
print("Original Generated Summary:")
print(generated_summary)

# Print enhanced summary with added information
print("\nEnhanced Summary with added information:")
print(enhanced_summary)

# Print added sentences at the bottom
print("\nAdded Sentences:")
for sentence in added_sentences:
    print(sentence)
```

```python
# Function to generate template sentences for missing annotations and enhance HPI
def enhance_summary_with_missing_annotations(transcript_annotations_df, generated_summary):
    # Generate template sentences based on missing annotations
    sample_sentence = "The patient reports taking {} of {} {}."

    # Convert generated summary to a list of words
    generated_summary_words = generated_summary.split()
    additional_sentences_words = []

    # Initialize a list to store added sentences
    added_sentences = []

    # Iterate through transcript annotations to find drug, strength, and frequency
    drugs = transcript_annotations_df[transcript_annotations_df["Label"] == "DRUG"]["Text"].tolist()
    strengths = transcript_annotations_df[transcript_annotations_df["Label"] == "STRENGTH"]["Text"].tolist()
    frequencies = transcript_annotations_df[transcript_annotations_df["Label"] == "FREQUENCY"]["Text"].tolist()

    # Check if we have annotations for all three: drug, strength, and frequency
    if drugs and strengths and frequencies:
        for drug, strength, frequency in zip(drugs, strengths, frequencies):
            # Format the sample sentence with the appropriate values
            additional_sentence = sample_sentence.format(strength, drug, frequency)
            added_sentences.append(additional_sentence)
            additional_sentences_words.extend(additional_sentence.split())

    # Concatenate the lists of words
    enhanced_summary_words = generated_summary_words + additional_sentences_words

    # Join the words back into sentences
    enhanced_summary = ' '.join(enhanced_summary_words)

    return enhanced_summary, added_sentences

# Call the function to enhance generated summary
enhanced_summary, added_sentences = enhance_summary_with_missing_annotations(transcript_annotations_df, generated_summary)

# Print original generated summary
print("Original Generated Summary:")
print(generated_summary)

# Print enhanced summary with added information
print("\nEnhanced Summary with added information:")
print(enhanced_summary)

# Print added sentences
print("\nAdded Sentences:")
for sentence in added_sentences:
    print(sentence)
```
